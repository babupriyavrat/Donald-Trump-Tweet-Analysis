```python
def preprocess_eng(text):
    #Create a clone of my repository and copy the stop_words folder in site_packages/stop-words
    #https://github.com/babupriyavrat/stop-words
    cleantext = BeautifulSoup(text, "lxml").text
    tokenizer = RegexpTokenizer(r'\w+')
    raw = cleantext.lower()
    tokens = tokenizer.tokenize(raw)
    
        # Remove numbers, but not words that contain numbers.
    tokens=[token for token in tokens if not token.isdigit()] 
    
    # Remove words that are only one character.
    tokens=[token for token in tokens if len(token) > 3]
    
    # create eng stop words list
    eng_stop_words = get_stop_words('en')
    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in eng_stop_words]
    #custom stop words
    nonsense_list=['https','http','www','facebook','twitter']
    stopped_tokens = [i for i in stopped_tokens if not i in nonsense_list]
    
    # Lemmatize all words in documents.
    lemmatizer = WordNetLemmatizer()
    stopped_tokens = [lemmatizer.lemmatize(token) for token in stopped_tokens]
  
    itemarr=[]
    for item in stopped_tokens:
        itemarr.append(item)
    return itemarr
```

```python
def save_dict(di_, filename_):
    with open(filename_, 'wb') as f:
        pickle.dump(di_, f)

def load_dict(filename_):
    with open(filename_, 'rb') as f:
        ret_di = pickle.load(f)
    return ret_di
```

```python
def generateLDA(pickle_name,num_topics):
    tweet_dict=load_dict(pickle_name)
    tweets= list(tweet_dict.values())
    
    from gensim.models import Phrases
    # Add bigrams and trigrams to tweets (only ones that appear 10 times or more).
    bigram = Phrases(tweets, min_count=4)
    trigram = Phrases(bigram[tweets])

    for idx in range(len(tweets)):
        for token in bigram[tweets[idx]]:
            if '_' in token:
                # Token is a bigram, add to document.
                tweets[idx].append(token)
        for token in trigram[tweets[idx]]:
            if '_' in token:
                # Token is a bigram, add to document.
                tweets[idx].append(token)
            
    from gensim import corpora, models
     # turn our tokenized documents into a id <-> term dictionary
    dictionary = corpora.Dictionary(tweets)
    import gensim  
    import bigfloat
    bigfloat.exp(5000,bigfloat.precision(100))
    random_seed = 69
    state = np.random.RandomState(random_seed) 
    start_time = time.time()
    corpus = [dictionary.doc2bow(tweet) for tweet in tweets]
    #print(corpus)
    corpora.MmCorpus.serialize('donaldtrumptweet.corpus.mm', corpus)
    print("start generating LDA model:")
    ldaModel = gensim.models.ldamodel.LdaModel(corpus, num_topics, id2word=dictionary,random_state=state)
    print("time took to generate LDA Model: --- %s seconds ---" % (time.time() - start_time))
    #save the dictionary and ldaModel
    dictionary.save("donaldtweet.dictionary.saved.lda")
    ldaModel.save('donaldtweet.model.lda')
    
    from gensim.similarities import MatrixSimilarity
    index = MatrixSimilarity(ldaModel[corpus])
    index.save('donaldtweet.lda.simIndex.index')
    return corpus,ldaModel,dictionary
        
```

```python
LDACorpus,LDAModel,LDADictionary=generateLDA('donald_trump_tweet_dict.pkl',100)
```

```python
def ldaSimilarity(pickle_name,dictionary,ldamodel,timestamp,n_tweets):
    similar_tweets={}
    tweet_dict=load_dict(pickle_name)
    vec_bow = dictionary.doc2bow(tweet_dict[str(timestamp)])
    vec_lda = ldamodel[vec_bow]
    index = utl.SaveLoad.load("donaldtweet.lda.simIndex.index")
    sims = index[vec_lda]
    tweet_timestamp= list(tweet_dict.keys())
    sims = sorted(enumerate(sims), key=lambda item: -item[1])
    sim_dict=dict(sims[:n_tweets])
    iterator=1
    for key, value in sim_dict.items():
        df2=df[df['date']==tweet_timestamp[key]]
        similar_tweets[iterator]=df2.loc[:,['date','retweets','favorites','text']].to_dict(orient="index")
        iterator=iterator+1
    return similar_tweets
    
    
```
